```python
plt.figure(figsize=(20,4))
ax=sns.countplot(data=crop_data2,x="State_Name")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(20,4))
ax=sns.countplot(data=crop_data2,x="Season")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(20,4))
ax=sns.countplot(data=crop_data2,x="Crop_Year")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb=crop_data2.groupby("State_Name").agg({"Area":"mean"})
print(gb)
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(gb,annot=True)
plt.show()
```

```python
gb=crop_data2.groupby("Season").agg({"Production":"mean"})
print(gb)
```

```python
plt.figure(figsize=(10,4))
sns.heatmap(gb,annot=True)
plt.show()
```
