```python
gb=amazon_data2.groupby("Item Type").agg({"Units Sold":"mean","Unit Price":"mean","Unit Cost":"mean"})
print(gb)
```

```python
plt.figure(figsize=(10,4))
sns.heatmap(gb,annot=True)
plt.show()
```

```python
gb=amazon_data2.groupby("Country").agg({"Total Revenue":"mean","Total Cost":"mean","Total Profit":"mean"})
print(gb)
```

```python
plt.figure(figsize=(7,5))
ax=sns.countplot(data=amazon_data2,x="Country")
ax.bar_label(ax.containers[0])
plt.title("Country Wise Distribution")
plt.show()
```

```python
region1=amazon_data2.loc[(amazon_data2["Region"]=="Asia")].count()
print(region1)
```

```python
regionA=amazon_data2.loc[(amazon_data2["Region"]=="Australia and Oceania")].count()
regionB=amazon_data2.loc[(amazon_data2["Region"]=="Central America And Caribbean")].count()
regionC=amazon_data2.loc[(amazon_data2["Region"]=="Europe")].count()
regionD=amazon_data2.loc[(amazon_data2["Region"]=="Sub-Saharan Africa")].count()
regionE=amazon_data2.loc[(amazon_data2["Region"]=="Asia")].count()
regionF=amazon_data2.loc[(amazon_data2["Region"]=="Middle East and North Africa")].count()
regionG=amazon_data2.loc[(amazon_data2["Region"]=="North Amrica")].count()
mlist=[regionA["Region"],regionB["Region"],regionC["Region"],regionD["Region"],regionE["Region"],regionF["Region"],regionG["Region"]]
l= ["regionA","regionB","regionC","regionD","regionE","regionF","regionG"]
plt.pie(mlist,labels=l)
```

```python
plt.title("DISTRIBUTION OF REGION")
plt.pie(mlist,labels=l)
plt.show()
```
